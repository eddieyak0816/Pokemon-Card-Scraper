```python
from pokemontcgsdk import Card
from pokemontcgsdk import Set
from time import sleep
import urllib
import requests
import cv2
import os
import pandas as pd
```

```python
setNames = [set.name for set in Set.all()]
len(setNames)
```

```python
for i, setname in enumerate(setNames):
    if os.path.exists(os.path.join('csv_new', setname.replace(':', '')+'.csv')):
        continue
    else:
        print(setname)
```

```python
cols = ["Image_Names", "Card_ID", "Card_Name", "Super_Type", "Types", "Set_Name", "Number", "Rarity", "Large_Image_URL", "Small_Image_URL", 
        "Normal_low", "Normal_mid", "Normal_high", "Normal_market", "Normal_directLow",
        "Holofoil_low", "Holofoil_mid", "Holofoil_high", "Holofoil_market", "Holofoil_directLow",
        "ReverseHolofoil_low", "ReverseHolofoil_mid", "ReverseHolofoil_high", "ReverseHolofoil_market", "ReverseHolofoil_directLow",
        "FirstEditionHolofoil_low", "FirstEditionHolofoil_mid", "FirstEditionHolofoil_high", "FirstEditionHolofoil_market", "FirstEditionHolofoil_directLow",
        "FirstEditionNormal_low", "FirstEditionNormal_mid", "FirstEditionNormal_high", "FirstEditionNormal_market", "FirstEditionNormal_directLow"]
```

```python
def downImg(image_url, setName, num):
    if not os.path.exists(os.path.join("images", setName)):
        os.mkdir(os.path.join("images", setName))
    
    img_data = requests.get(image_url).content
    with open('images/{}/{:03d}.jpg'.format(setName, num), 'wb') as handler:
        handler.write(img_data)
```

```python
for j, setname in enumerate(setNames):
    if setname != 'Perfect Order':
        continue
    print(setname, end = " " )
    setCards = Card.where(q="set.name:{}".format(setname.replace(" ", "\ ").replace(':', '\:').replace('&', '\&').replace('—', "\—").replace("\'s", "\\'s")))
    print(": " + str(len(setCards)))
    
    details = []
    for i in range(len(setCards)):
        setCard = setCards[i]
        # print(i, end=", ")
        # print(setCard)
        # print()
        # print()
        # continue
        L_imgURL = setCard.images.large
        S_imgURL = setCard.images.small
        if L_imgURL is not None:
            downImg(L_imgURL, setname, i)
        elif S_imgURL is not None:
            downImg(S_imgURL, setname, i)

        #///////////////////////
        imgName = '{:03d}.jpg'.format(i)
        CardID = setCard.id
        CardName = setCard.name
        superType = setCard.supertype
        Types = setCard.types
        if Types is not None:
            Types = ("__").join(Types)
        SetName = setCard.set.name
        Number = setCard.number
        Rarity = setCard.rarity
        
        if setCard.tcgplayer is None or setCard.tcgplayer.prices is None:
            prices = [None]*25
        
        else:
            prices = []
            
            ###
            obj = setCard.tcgplayer.prices.normal
            if obj is None:
                p = [None]*5
            else:
                p = [obj.low, obj.mid, obj.high, obj.market, obj.directLow]
            prices += p
            
            ###
            obj = setCard.tcgplayer.prices.holofoil
            if obj is None:
                p = [None]*5
            else:
                p = [obj.low, obj.mid, obj.high, obj.market, obj.directLow]
            prices += p
            
            ###
            obj = setCard.tcgplayer.prices.reverseHolofoil
            if obj is None:
                p = [None]*5
            else:
                p = [obj.low, obj.mid, obj.high, obj.market, obj.directLow]
            prices += p
            
            ###
            obj = setCard.tcgplayer.prices.firstEditionHolofoil
            if obj is None:
                p = [None]*5
            else:
                p = [obj.low, obj.mid, obj.high, obj.market, obj.directLow]
            prices += p
            
            ###
            obj = setCard.tcgplayer.prices.firstEditionNormal
            if obj is None:
                p = [None]*5
            else:
                p = [obj.low, obj.mid, obj.high, obj.market, obj.directLow]
            prices += p
            
            
        #################
        cardDetails = [imgName, CardID, CardName, superType, Types, SetName, Number, Rarity, L_imgURL, S_imgURL] + prices
        
        details.append(cardDetails)
        
        
    #-----------------------
    df = pd.DataFrame(details, columns=cols)
    df.to_csv("csv_new/" + setname.replace(':', '') + ".csv", index=False)
    
    if (j+1)%5 == 0:
        print("Sleeping for 50 sec.")
        sleep(50)
```
